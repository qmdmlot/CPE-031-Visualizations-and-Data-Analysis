```python
#rep is Jones
df3 = df2.loc[(df2['Rep']=="Jones") & (df2['Region'] == 'East')]
display(df3)
```

```python
day_unit_sales = df3.groupby('OrderDate')['Units'].sum()
plt.figure(figsize=(12,6))
plt.plot(day_unit_sales.index, day_unit_sales.values)
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.title('Units Sold by Jones in the Eastern Region by Date')
plt.xticks(rotation=45)
plt.show()
```
